```python
%matplotlib inline
```

```python
import thinkbayes
import thinkplot
```

```python
class Dice(thinkbayes.Pmf):
    def __init__(self, num_sides):
        thinkbayes.Pmf.__init__(self)
        for i in xrange(1, num_sides+1):
            self.Set(i, 1)
        self.Normalize()
```

```python
d6 = Dice(6)
```

```python
three = thinkbayes.SampleSum([d6] * 3, 5000)
three.Mean()
```

```python
three_exact = d6+d6+d6

three_exact.Mean()
```

```python
thinkplot.PrePlot(2)
thinkplot.Pmf(three)
thinkplot.Pmf(three_exact, linestyle='dashed')
```

```python
thinkplot.Pmf(three.Max(3).MakePmf())
```
