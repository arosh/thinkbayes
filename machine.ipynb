```python
from thinkbayes import Suite
class Machine(Suite):
    def __init__(self, y, values=None, name=''):
        Suite.__init__(self, values, name)
        self.y = y
    def Likelihood(self, data, hypo):
        if data == 'H':
            return ((hypo) * (100 - self.y) + (100 - hypo) * self.y) / 100.0
        else:
            return ((100 - hypo) * (100 - self.y) + hypo * self.y) / 100.0
```

```python
import thinkplot
for y in xrange(0, 11):
    suite = Machine(y=y*10, values=xrange(101))
    suite.UpdateSet('H' * 140)
    suite.UpdateSet('T' * 110)
    thinkplot.Pmf(suite)
```
